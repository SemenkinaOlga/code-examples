###### Info: Get API key form OMDb: https://www.omdbapi.com/

```python
import requests
from IPython.display import Image
import json
import re
```

```python
f = open("api_key.txt", "r")
api_key = f.read()
```

## Find movie by name and year

```python
title = "The Game"
year = 1997
```

```python
url = "http://www.omdbapi.com/?apikey="+api_key+"&t="+title.replace(" ", "+")+"&y="+str(year)+"&plot=full"
headers = {"accept": "application/json"}
response = requests.get(url, headers=headers)
```

```python
if response.status_code == 200:
    print ('Ok')
else:
    print ('Error')
```

```python
result = json.loads(response.text)
result
```

```python
image_url = result["Poster"]
```

```python
Image(requests.get(image_url).content)
```

## Fing movie by imdb id

```python
imdbID = "tt17491088"
```

```python
url = "http://www.omdbapi.com/?apikey=" + api_key + "&i=" + imdbID + "&plot=full"
headers = {"accept": "application/json"}
response = requests.get(url, headers=headers)
if response.status_code == 200:
    print ('Ok')
else:
    print ('Error')
```
